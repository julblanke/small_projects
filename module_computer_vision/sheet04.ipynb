### a) Histogram-based segmentation

What is histogram-based segmentation? What are it's goals, benefits, and problems?

Histogram-based segmentation tries to find different segments by computing the histogram of the color value and to label entities as a segment that are seperated by different thresholds in the histogram. \
With this method one can try to extract the background and the foreground (or more entities in the foreground) of the image. This method is easy to do and has a low complexity. \
Problems can occur, when (i.e.) the background is not evenly colored. Shadows on the image can make histogram-based segmentation fail.

## Assignment 2: Pyramid representation [5 Points]

**a)** What is the *Gaussian pyramid*? How does the **reduce** operation work? Explain in your own words what low pass filtering is and why it should be used when building the pyramid? Implement the **reduce** operation and generate a figure similar to the one on (CV-07 slide 32).

The base of a gaussian pyramid is an image. On each higher level of the pyramid, a new picture is created which is 1/4th the size of the lower level. Each pixel at level $i$ replaces 4 pixels at level $i-1$. This is the reduce operation. \
Low pass filtering means that the image is filtered before the reduce operation is called, so that high frequencies are removed to avoid artifacts.

```python
%matplotlib inline
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from imageio import imread

img = imread('images/mermaid.png')

pyramid_image = img.copy() # change this!

# YOUR CODE HERE
pyramid = [pyramid_image]
while not (pyramid_image.shape[0] <= 1 and pyramid_image.shape[1] <= 1):
    
    # low pass filtering
    binom_filter = [[1/16, 2/16, 1/16],
                    [2/16, 4/16, 2/16],
                    [1/16, 2/16, 1/16]]
    
    pyramid_image = ndimage.convolve(pyramid_image, binom_filter)
    
    # initialize the higher-level image with 1/4th size
    tmp_image = np.ndarray((int(pyramid_image.shape[0] / 2), int(pyramid_image.shape[1] / 2)))
    
    if tmp_image.shape[0] == 0 or tmp_image.shape[1] == 0:
        tmp_image = np.ndarray((1,1))
    
    # compute the pixel values in the higher level
    for i in range(tmp_image.shape[0]):
        for j in range(tmp_image.shape[1]):
            h = pyramid_image.shape[0]
            w = pyramid_image.shape[1]
            
            # take the mean value of 4 pixels
            tmp_image[i][j] = (int(pyramid_image[(i * 2) % h][(j * 2) % w])
                               + int(pyramid_image[(i * 2 + 1) % h][(j * 2) % w])
                               + int(pyramid_image[(i * 2) % h][(j * 2 + 1) % w])
                               + int(pyramid_image[(i * 2 + 1) % h][(j * 2 + 1) % w])) / 4
    
    pyramid_image = tmp_image
    pyramid.append(pyramid_image)
    
# Create an image similiar to CV-07 slide 32
img_pyramid = np.ndarray((img.shape))

y_offset = 0

for level in pyramid:
    for i in range(level.shape[0]):
        for j in range(level.shape[1]):
            img_pyramid[i + y_offset][j] = level[i][j]
    y_offset = pyramid[0].shape[0] - int(level.shape[0] / 2) - 1

pyramid_image = img_pyramid
    
plt.figure(figsize=(15,10))
plt.gray()
plt.imshow(pyramid_image)
plt.show()
```

**b)** What is a co-occurence matrix? How can it be used to characterize texture?

The co-occurence matrix represents the correlation between pixels. The color values of two pixels with a specific distance on a line with a given angle to the x-axis are compared. \
The idea is that different textures produce different co-occurence matrices and therefore textures can be differentiated by their matrices.


**c)** Implement a function to compute the co-occurence matrix of an image (patch). Apply it and compare your results to (CV-07 slide 54).

```python
%matplotlib inline
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import imageio 

img = imageio.imread('images/mermaid.png')#, mode='L')

def get_patch(img, x, y, size=40):
    """
    Extract a rectangular patch from an image and mark it in the original image.
    
    Args:
        img (nndarray): Input image.
        x (uint): X-coordinate.
        y (uint): Y-coordinate.
        size (uint): Size of the patch.
        
    Returns:
        result: The extracted patch.
    """
    result = img[x:x+size,y:y+size].copy()
    img[x:x+size, [y,y+1,y+size,y+size+1]] = 0
    img[[x,x+1,x+size,x+size+1], y:y+size] = 0
    return result

patches = []
patches.append(get_patch(img, 50,130))
patches.append(get_patch(img, 110,80))
patches.append(get_patch(img, 260,340))
patches.append(get_patch(img, 310,110))
patches.append(get_patch(img, 100,440))


def cooccurrence(img, dx=1, dy=1):
    """
    Compute a co-occurence matrix for the given image.
    
    Args:
        img          the grayscale image (uint8)
        dx,dy        the offset between the two reference points

    Returns:
        matrix       the co-occurence matrix
    """
    matrix = np.empty((256, 256))
    # YOUR CODE HERE
    
    # initialize the matrix with zeros, otherwise the values are 'empty', which are really small values that produce
    # a runtime warning and random outputs
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            matrix[i][j] = 0
    
    
    # normalization factor
    n = 1 / (img.shape[0] * img.shape[1])
    
    # iterate through every pixel in the image
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            
            # skip if index is out of bound
            if (i + dx < 0) or (i + dx >= img.shape[0]) or (j + dy < 0) or (j + dy >= img.shape[1]):
                continue
            
            # we don't have to iterate through every possible pixel-value combination of g_1 and g_2
            # the only 1-value is returned if (value pixel) = g_1 and (value pixel+offset) = g_2, so the value in the
            # co_matrix at position (value pixel, value pixel+offset) can be incremented
            # This saves a lot of computation time
            matrix[img[i][j]][img[i + dx][j + dy]] += (n * 1)
                    
    return matrix


plt.figure(figsize=(8, 8))
plt.gray()
plt.imshow(img)
plt.show()


plt.figure(figsize=(8, 8))
i = 0
for p in patches:
    plt.subplot(len(patches),3,i+1); plt.axis('off'); plt.imshow(p)
    # For visualization one may apply some extra me, e.g., logarithmization or binarization
    plt.subplot(len(patches),3,i+2); plt.imshow(np.log(1 + cooccurrence(p, 0, 1)), interpolation='none')
    #plt.subplot(len(patches),3,i+2); plt.imshow(cooccurrence(p, 0, 1)>0, interpolation='none')
    plt.subplot(len(patches),3,i+3); plt.imshow(cooccurrence(p,1,0)>0, interpolation='none')
    i += 3
plt.show()
```

## Assignment 4: Region merging [5 Points]

Implement the *region merging* algorithm (CV-07 slide 39) and apply it to the image `segments.png` (or some part of it). Use a simple *homogeneity condition*, e.g. that the maximal difference between gray values in a segment is not larger than a given threshold.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import imageio
import warnings
from operator import add
from functools import reduce
warnings.filterwarnings('ignore')

img = imageio.imread('./images/segments.png', pilmode='L')
# Choosing a large image region lengthens computation time
img = img[64:128,64:128]

# compute the `label` array by implementing "region merging"
# YOUR CODE HERE

thresh = 50

def split(image):
    half_split = np.array_split(image, 2)
    res = map(lambda x: np.array_split(x, 2, axis=1), half_split)
    return reduce(add, res)

def concatenate(north_west, north_east, south_west, south_east):
    top = np.concatenate((north_west, north_east), axis=1)
    bottom = np.concatenate((south_west, south_east), axis=1)
    return np.concatenate((top, bottom), axis=0)

def calculate_mean(img):
    return np.mean(img, axis=(0, 1))

def check_thresh(img, thresh):
    if (max(img.flatten())-min(img.flatten())) < thresh:
        return True
    else:
        return False

class QuadTree:
    
    def insert(self, img, level = 0):
        self.level = level
        self.mean = calculate_mean(img).astype(int)
        self.resolution = (img.shape[0], img.shape[1])
        self.final = True
        
        if not check_thresh(img, thresh):
            split_img = split(img)
            
            self.final = False
            self.north_west = QuadTree().insert(split_img[0], level + 1)
            self.north_east = QuadTree().insert(split_img[1], level + 1)
            self.south_west = QuadTree().insert(split_img[2], level + 1)
            self.south_east = QuadTree().insert(split_img[3], level + 1)

        return self
    
    def get_image(self, level):
        if(self.final or self.level == level):
            return np.tile(self.mean, (self.resolution[0], self.resolution[1], 1))
        
        return concatenate(
            self.north_west.get_image(level), 
            self.north_east.get_image(level),
            self.south_west.get_image(level),
            self.south_east.get_image(level))
    
quadtree = QuadTree().insert(img)
    
plt.figure(figsize=(12, 12))
plt.gray()
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(quadtree.get_image(1000)[:, :, 0], cmap='gray')
plt.show()

warnings.filterwarnings('ignore')
```
